```python
import pickle
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

# Arrays to store object points and image points from all the images
objpoints = [] # 3D points in real world space
imgpoints = [] # 2D points in image plane
```

```python
# Prepare object points, like (0,0,0), (1,0,0)...
objp = np.zeros((6 * 8, 3), np.float32)
objp[:,:2] = np.mgrid[0:8, 0:6].T.reshape(-1, 2) # x, y coordinates
img = mpimg.imread('./camera_cal/calibration2.jpg')
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# Find the chessboard grayscale
ret, corners = cv2.findChessboardCorners(gray, (8,6), None)
plt.imshow(gray)
# plt.show()
```

```python
# If corners are found, add objects points, image points
if ret:
    imgpoints.append(corners)
    objpoints.append(objp)
    
    # draw and display the corners
    img = cv2.drawChessboardCorners(img, (8,6), corners, ret)
    plt.imshow(img)
    plt.show()
```

```python
import os

list_dir = np.sort(os.listdir('camera_cal'))
```

```python
print (list_dir)
```

```python
for image in list_dir:
    img = mpimg.imread('camera_cal/' + image)
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    ret, corners = cv2.findChessboardCorners(gray, (8,6), None)
    
    if ret:
        imgpoints.append(corners)
        objpoints.append(objp)
        
        img = cv2.drawChessboardCorners(img, (8,6), corners, ret)
        plt.imshow(img)
```

```python
print (np.array(imgpoints).shape)
print (np.array(objpoints).shape)
```

```python
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, gray.shape[::-1],None,None)
```

```python
img = mpimg.imread('camera_cal/calibration2.jpg')
h,  w = img.shape[:2]
newcameramtx, roi=cv2.getOptimalNewCameraMatrix(mtx,dist,(w,h),1,(w,h))
```

```python
# undistort
dst = cv2.undistort(img, mtx, dist, None, newcameramtx)
# crop the image
x,y,w,h = roi
dst = dst[y:y+h, x:x+w]
cv2.imwrite('calibresult.png',dst)
# plt.imshow(dst)
```
